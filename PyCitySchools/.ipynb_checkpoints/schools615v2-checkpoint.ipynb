```python
import pandas as pd

schools= "resources/schools_complete.csv"

students="resources/students_complete.csv"

school_data = pd.read_csv(schools)

student_data = pd.read_csv(students)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#District Summary Calculations

total_schools = school_data['school_name'].count()

total_students = student_data['Student ID'].count()

total_budget = school_data['budget'].sum()

avg_math = student_data['math_score'].mean()

avg_read = student_data['reading_score'].mean()

pass_math = student_data.loc[student_data['math_score'] >= 70, ['math_score']]

percent_pass_math = pass_math["math_score"].count()/total_students*100

pass_read = student_data.loc[student_data['reading_score'] >= 70, ['reading_score']]

percent_pass_read = pass_read["reading_score"].count()/total_students*100

pass_overall = student_data.loc[(student_data["reading_score"] >= 70) & (student_data["math_score"] >= 70),:]

percent_pass_overall = pass_overall["student_name"].count()/total_students*100
```

```python
# DataFrame of summary calculations

district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget":[total_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_read], 
    "% Passing Math":[percent_pass_math],
    "% Passing Reading":[percent_pass_read],
    "Overall Passing Rate":[percent_pass_overall],
})

#format the dataframe

district_summary['Total Students'] = district_summary['Total Students'].astype(float).map("{:,.0f}".format)
district_summary['Total Budget'] = district_summary['Total Budget'].astype(float).map("${:,.0f}".format)
district_summary['Average Math Score'] = district_summary['Average Math Score'].astype(float).map("{:,.2f}".format)
district_summary['Average Reading Score'] = district_summary['Average Reading Score'].astype(float).map("{:,.2f}".format)
district_summary['% Passing Math'] = district_summary['% Passing Math'].astype(float).map("{:,.2f}%".format)
district_summary['% Passing Reading'] = district_summary['% Passing Reading'].astype(float).map("{:,.2f}%".format)
district_summary['Overall Passing Rate'] = district_summary['Overall Passing Rate'].astype(float).map("{:,.2f}%".format)


district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:

  * % Passing Math

* Create a dataframe to hold the above results

```python
#School Summary Calculations

school_name = school_data['school_name']

school_type = school_data['type']

size = school_data['size']

total_students_per_school = student_data['school_name'].value_counts()

#total school budget is total_budget calculated above

per_student_budget = school_data['budget']/size

```

```python
#Group By School calculations

grouped_school_name = student_data.groupby(['school_name']) 

avg_math_by_school = grouped_school_name['math_score'].mean() 

avg_read_by_school = grouped_school_name['reading_score'].mean() 

#Create Dataframe with school average scores

df_of_averages = pd.DataFrame({"Average Math Score": avg_math_by_school,
                              "Average Reading Score": avg_read_by_school,}).reset_index()

# df with students Passing Math

students_math_passing = student_data.loc[student_data['math_score'] >= 70]

all_students_math_passing= pd.DataFrame({"school_name": students_math_passing["school_name"],
                                       "Count_students_passing_math": 0})

students_pass_math_school = all_students_math_passing.groupby(['school_name']).count().reset_index()

# df with students Passing Reading

students_read_passing = student_data.loc[student_data['reading_score'] >= 70]

all_students_read_passing= pd.DataFrame({"school_name": students_read_passing["school_name"],
                                       "Count_students_passing_read": 0})

students_pass_read_school = all_students_read_passing.groupby(['school_name']).count().reset_index()


# df with students Passing Reading & Math

students_overall_passing = student_data.loc[(student_data['reading_score'] >= 70) & (student_data["math_score"] >= 70),:]

all_students_overall_passing= pd.DataFrame({"school_name": students_overall_passing["school_name"],
                                       "Count_students_passing_overall": 0})

students_pass_overall_school = all_students_overall_passing.groupby(['school_name']).count().reset_index()

#group students by school to calculate average math and reading scores per school
school_group = student_data.groupby(['school_name'])

# Create merge of school data with student data

new_school_df = pd.merge(school_data, df_of_averages, on='school_name')
new_school_df = pd.merge(new_school_df, students_pass_math_school, on = 'school_name')
new_school_df = pd.merge(new_school_df, students_pass_read_school, on = 'school_name')
new_school_df = pd.merge(new_school_df, students_pass_overall_school, on = 'school_name')


#calculate percent of students passing math
percent_passing_math_school = (new_school_df['Count_students_passing_math'] / new_school_df['size']) * 100

#calculate percentage of students passing read
percent_passing_reading_school = (new_school_df['Count_students_passing_read'] / \
                                      new_school_df['size']) * 100

#calculate overall passing rate
overall_passing_school = (new_school_df['Count_students_passing_overall'] / \
                                      new_school_df['size']) * 100




# create district_summary dataframe for display
schools_summary_df = pd.DataFrame({'School Name': new_school_df['school_name'],
                               'School Type': new_school_df['type'],
                               'Total Students': new_school_df['size'],
                               'Total School Budget': new_school_df['budget'],
                               'Per Student Budget': (new_school_df['budget'] / new_school_df['size']),
                               'Average Math Score': new_school_df['Average Math Score'],
                               'Average Reading Score': new_school_df['Average Reading Score'],
                               '% Passing Math': percent_passing_math_school,
                               '% Passing Reading':percent_passing_reading_school,
                               'Overall Passing Rate': overall_passing_school})

# schools_summary_df['Total Students'] = schools_summary_df['Total Students'].astype(float).map("{:,.0f}".format)
# schools_summary_df['Total School Budget'] = schools_summary_df['Total School Budget'].astype(float).map("{:,.0f}".format)
# schools_summary_df['Average Math Score'] = schools_summary_df['Average Math Score'].astype(float).map("{:,.2f}".format)
# schools_summary_df['Average Reading Score'] = schools_summary_df['Average Reading Score'].astype(float).map("{:,.2f}".format)
# schools_summary_df['% Passing Math'] = schools_summary_df['% Passing Math'].astype(float).map("{:,.2f}%".format)
# schools_summary_df['% Passing Reading'] = schools_summary_df['% Passing Reading'].astype(float).map("{:,.2f}%".format)
# schools_summary_df['Overall Passing Rate'] = schools_summary_df['Overall Passing Rate'].astype(float).map("{:,.2f}%".format)


school_summary_df = schools_summary_df[['School Name', 'School Type', 'Total Students', 'Total School Budget',
                                 'Per Student Budget', 'Average Math Score', 'Average Reading Score',
                                 '% Passing Math', '% Passing Reading', 'Overall Passing Rate']] \
                 .set_index('School Name').rename_axis(None)

# school_summary_df.describe()
school_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Groupby high school and grade

avg_scores_by_high_school_df = student_data.groupby(['school_name', 'grade']).mean().reset_index()



#individual grades
grade9_df = avg_scores_by_high_school_df.loc[avg_scores_by_high_school_df['grade'] == '9th']
grade10_df = avg_scores_by_high_school_df.loc[avg_scores_by_high_school_df['grade'] == '10th']
grade11_df = avg_scores_by_high_school_df.loc[avg_scores_by_high_school_df['grade'] == '11th']
grade12_df = avg_scores_by_high_school_df.loc[avg_scores_by_high_school_df['grade'] == '12th']

#identify columns to include in df
nine_df = pd.DataFrame({"school_name": grade9_df["school_name"],
                                   "reading_score_9th": grade9_df['reading_score'],
                                   "math_score_9th": grade9_df['math_score']})
ten_df = pd.DataFrame({"school_name": grade10_df["school_name"],
                                   "reading_score_10th": grade10_df['reading_score'],
                                   "math_score_10th": grade10_df['math_score']})
eleven_df = pd.DataFrame({"school_name": grade11_df["school_name"],
                                   "reading_score_11th": grade11_df['reading_score'],
                                   "math_score_11th": grade11_df['math_score']})
twelve_df = pd.DataFrame({"school_name": grade12_df["school_name"],
                                   "reading_score_12th": grade12_df['reading_score'],
                                   "math_score_12th": grade12_df['math_score']})

all_scores_by_grade_hs_df = pd.merge(nine_df, ten_df, on = 'school_name')\
.merge(eleven_df, on = 'school_name')\
.merge(twelve_df, on = 'school_name') 
            

math_scores_by_grade_df = pd.DataFrame({'school_name': all_scores_by_grade_hs_df['school_name'],
                                        '9th': all_scores_by_grade_hs_df['math_score_9th'],
                                        '10th': all_scores_by_grade_hs_df['math_score_10th'],
                                        '11th': all_scores_by_grade_hs_df['math_score_11th'],
                                        '12th': all_scores_by_grade_hs_df['math_score_12th']})

math_scores_by_grade_df = math_scores_by_grade_df[['school_name', '9th', '10th', '11th', '12th']]\
                 .set_index('school_name').rename_axis(None)
math_scores_by_grade_df
```

```python
#now we need to create another dataframe with reading scores
avg_read_scores_by_high_school_df = pd.DataFrame({'school_name': all_scores_by_grade_hs_df['school_name'],
                                        '9th': all_scores_by_grade_hs_df['reading_score_9th'],
                                        '10th': all_scores_by_grade_hs_df['reading_score_10th'],
                                        '11th': all_scores_by_grade_hs_df['reading_score_11th'],
                                        '12th': all_scores_by_grade_hs_df['reading_score_12th']})

#Reassemble columns in display order, set index to school and delete index label
avg_read_scores_by_high_school_df = avg_read_scores_by_high_school_df[['school_name', '9th', '10th', '11th', '12th']]\
                 .set_index('school_name').rename_axis(None)
    
#display dataframe
avg_read_scores_by_high_school_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#create a copy to save the old data frame

school_spending = school_summary_df.copy()

#create bins to hold values
bins = [575, 585, 595, 605, 615, 625, 650]

#name the bins
view_groups = ['\$575-\$585','\$585-\$595','\$595-\$605','\$605-\$615','\$615-\$625', '>$625']

view_groups

#Convert student budget into a float
school_spending['Per Student Budget']=school_spending['Per Student Budget']\
.replace('[\$]','',regex=True).astype('float')


#append a new column that shows the spend ranges
school_spending['Spending Ranges Per Student'] = pd.cut(school_spending['Per Student Budget'],bins, labels=view_groups)


school_spending

#build the dataset and include the column we created above
school_spending = school_spending[['Average Math Score',
                                            'Average Reading Score',
                                            '% Passing Math',
                                            '% Passing Reading',
                                            'Overall Passing Rate',
                                            'Spending Ranges Per Student']]

#group by Spending Range
average_scores_by_spending_df = school_spending.groupby('Spending Ranges Per Student').mean()

# average_scores_by_spending_df['Average Math Score'] = average_scores_by_spending_df['Average Math Score'].astype(float).map("{:,.0f}".format)
# average_scores_by_spending_df['Average Reading Score'] = average_scores_by_spending_df['Average Reading Score''].astype(float).map("{:,.0f}".format)
# average_scores_by_spending_df['% Passing Math'] = average_scores_by_spending_df['% Passing Math'].astype(float).map("{:,.0f}".format)
# average_scores_by_spending_df['% Passing Reading'] = average_scores_by_spending_df['% Passing Reading'].astype(float).map("{:,.0f}".format)
# average_scores_by_spending_df['Overall Passing Rate'] = average_scores_by_spending_df['Overall Passing Rate'].astype(float).map("{:,.0f}".format)
# average_scores_by_spending_df['Spending Ranges Per Student'] = average_scores_by_spending_df['Spending Ranges Per Student'].astype(float).map("{:,.0f}".format)

average_scores_by_spending_df
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#copy the data frame to save the old one
school_summary_bysize = school_summary_df.copy()

#create bins
bins = [0, 1000, 3000, 5000]

#name the bins
view_groups = ['Small (<1000) Students', 'Medium (1000-3000) Students', 'Large (3000-5000) Students']

#convert school size to numeric values
school_summary_bysize['Total Students'] = pd.to_numeric(school_summary_bysize['Total Students'])

#Cut the school size into the bins we define above 
school_summary_bysize['School Size'] = pd.cut(school_summary_bysize['Total Students'],
                                                                  bins, labels=view_groups)

#selecte columns for dataframe
school_summary_bysize = school_summary_bysize[['Average Math Score', 'Average Reading Score',
                                          '% Passing Math', '% Passing Reading', 'Overall Passing Rate',
                                          'School Size']]

#group by size bins and and take the aveage of scores
scores_by_school_bysize_df = school_summary_bysize.groupby('School Size').mean()


scores_by_school_bysize_df
```
