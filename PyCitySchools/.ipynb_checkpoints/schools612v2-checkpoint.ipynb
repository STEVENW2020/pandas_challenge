```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
district = school_data_complete.groupby(['school_name'])
schoolcount = len(district)
totstudents = student_data["student_name"].count()
totbudget = school_data["budget"].sum()
perstubudget = totbudget/totstudents
avgmathscore = student_data["math_score"].mean()
avgreadscore = student_data["reading_score"].mean()

```

```python
goodmathstudents = student_data.loc[student_data["math_score"] >= 70,:] 
goodmathpercent = goodmathstudents["student_name"].count()/totstudents
# goodmathpercent
goodreadingstudents = student_data.loc[student_data["reading_score"] >= 70,:]
goodreadpercent = goodreadingstudents["student_name"].count()/totstudents
# goodreadpercent
beststudents = student_data.loc[(student_data["reading_score"] >= 70) & (student_data["math_score"] >= 70),:]
beststudentspercent = beststudents["student_name"].count()/totstudents
# beststudentspercent
```

```python
# Create a dataframe to hold the results of above

# print(schoolcount)
# print(totstudents)
# print(totbudget)
# print(perstubudget)
# print(avgmathscore)
# print(avgreadscore)
# print(goodmathpercent)
# print(goodreadpercent)
# print(beststudentspercent)


summary_df = pd.DataFrame({
    "Total Number of Schools":[schoolcount],
    "Total Number of Students":[totstudents],
    "Total Budget":[totbudget],
    "Total Budget per Student":[perstubudget],     
    "Average Math Score":[avgmathscore],
    "Average Reading Score":[avgreadscore],
    "Percent Passing Math":[goodmathpercent],
    "Percent Passing Reading":[goodreadpercent],
    "Percent Passing Math and Reading":[beststudentspercent] 
})
summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
studentcount = school_data_complete.groupby("school_name")["student_name"].count()
schoolbudget = school_data_complete.groupby("school_name")["budget"].sum()
schoolbudgetper = school_data_complete.groupby("school_name")["budget"].sum()/studentcount
schoolavgmath = school_data_complete.groupby("school_name")["math_score"].sum()/studentcount
schoolavgread = school_data_complete.groupby("school_name")["reading_score"].sum()/studentcount
schoolmathpass = school_data_complete.loc[student_data['math_score']>=70, :
                                 ].groupby('school_name')['math_score'].count()/studentcount

schoolreadpass = school_data_complete.loc[student_data['reading_score']>=70, :
                                 ].groupby('school_name')['reading_score'].count()/studentcount

schoolbothpass = school_data_complete.loc[(student_data['reading_score']>=70) & (student_data['math_score']>=70), :
                                 ].groupby('school_name')['reading_score'].count()/studentcount


# Count how many students by school and create DataFrame
district_df = pd.DataFrame(district)
# district_df.columns

schools_df = pd.concat([schoolbudgetper,
                        schoolavgmath, 
                        schoolavgread, 
                        schoolmathpass,
                        schoolreadpass,
                        schoolbothpass
                       ],axis=1)


schools_df.reset_index(inplace=False)


# schools_df

summary_schools = pd.merge(school_data, schools_df, how="outer", on=["school_name"])

summary_schools.reset_index(inplace=False)

summary_schools = summary_schools.rename(columns = {'school_name' : "School Name",
                                                    'type' : "School Type",
                                                    'size' : "Student Count",
                                                    'budget' : "School Budget",
                                                    0 : 'Budget per Student',
                                                    1 : 'Average Math Score',
                                                    2 : 'Average Reading Score',
                                                    3 : 'Percent Passing Math',
                                                    4 : 'Percent Passing Reading',
                                                    5 : 'Percent Passing Math & Reading',})

# summary_schools.set_index("School Name")

summary_schools
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
studentsgrade9 = student_data.loc[student_data['grade'] =='9th',]
studentsgrade10 = student_data.loc[student_data['grade'] =='10th',]
studentsgrade11 = student_data.loc[student_data['grade'] =='11th',]
studentsgrade12 = student_data.loc[student_data['grade'] =='12th',]

countgrade9 = studentsgrade9.groupby('school_name')['grade'].count()
countgrade10 = studentsgrade10.groupby('school_name')['grade'].count()
countgrade11 = studentsgrade11.groupby('school_name')['grade'].count()
countgrade12 = studentsgrade12.groupby('school_name')['grade'].count()

schoolgrade9M = studentsgrade9.groupby('school_name')['math_score'].sum()/countgrade9
schoolgrade10M = studentsgrade10.groupby('school_name')['math_score'].sum()/countgrade10
schoolgrade11M = studentsgrade11.groupby('school_name')['math_score'].sum()/countgrade11
schoolgrade12M = studentsgrade12.groupby('school_name')['math_score'].sum()/countgrade12

scoresummary = pd.concat([schoolgrade9M,
                           schoolgrade10M,
                           schoolgrade11M,
                           schoolgrade12M], axis=1)

scoresummary = scoresummary.rename(columns={0:'9th grade', 1:'10th grade', 2:'11th grade', 3:'12th grade'})

# del scoresummary.index.name

scoresummary
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
schoolgrade9r = studentsgrade9.groupby('school_name')['reading_score'].sum()/countgrade9
schoolgrade10r = studentsgrade10.groupby('school_name')['reading_score'].sum()/countgrade10
schoolgrade11r = studentsgrade11.groupby('school_name')['reading_score'].sum()/countgrade11
schoolgrade12r = studentsgrade12.groupby('school_name')['reading_score'].sum()/countgrade12

scoresummary2 = pd.concat([schoolgrade9r,
                           schoolgrade10r,
                           schoolgrade11r,
                           schoolgrade12r], axis=1)

scoresummary2 = scoresummary2.rename(columns={0:'9th grade', 1:'10th grade', 2:'11th grade', 3:'12th grade'})
scoresummary2
```
